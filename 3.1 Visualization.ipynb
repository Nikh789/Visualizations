# Matplotlib

```python
# MATPLOTLIB- package for 2D & 3D graphic library
# Benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals
# "matplotlib.pyplot" is a collection of command style functions that make matplotlib work like MATLAB

# SEABORN- used for advanced visualization graphs
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [2,4,6,8,10]
y = [2,4,9,6,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My Plot")
```

```python
plt.bar(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My Plot")
```

```python
plt.bar(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My Plot")
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My Plot")
```

```python
plt.pie(x)
```

```python
plt.pie(y)
```
