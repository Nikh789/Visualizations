```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
wbcd = pd.read_csv("C:\\Users\\nikhi\\OneDrive\\Desktop\\Social Prachar\\Excel files\\wbcd.csv")
wbcd
```

```python
del wbcd['id']
wbcd
```

```python
# To give the count of each category 
wbcd['diagnosis'].value_counts()
```

```python
freq = pd.crosstab(index = wbcd['diagnosis'], columns = "count")
freq
```

```python
freq/freq.sum()*100
```

```python
# To replace the function used to change the label name in the rows
wbcd['diagnosis'].replace({"B":"Benign","M":"Malignant"},inplace=True)
wbcd
```
